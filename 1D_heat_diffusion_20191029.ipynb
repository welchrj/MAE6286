1D Heat diffusion 

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
sigma = 0.5
l = 1.0
nx = 51
dx = l/(nx-1)
alpha = 1.22E-3
dt = sigma*dx**2/alpha
x = np.linspace(0.0, l, num=nx)
u0 = np.zeros(nx)
u0[0] = 100.0
```

```python
def ftcs(u0, sigma, nt):
    '''Forward in time central in space scheme for heat diffusion equation
    Dirichlet BC '''
    u = u0.copy()
    for i in range(nt):
        u = u.copy()
        u[1:-1] = u[1:-1] + sigma*(u[:-2] - 2*u[1:-1] + u[2:])
    return u 

def ftcs_neumann(u0, sigma, nt):
    '''FTCS with neumann conditions'''
    u = u0.copy()
    for i in range(nt):
        u = u.copy()
        u[1:-1] = u[1:-1] + sigma*(u[:-2] - 2*u[1:-1] + u[2:])
        u[-1] = u[-2]
    return u
```

```python
nt = 100
ans = ftcs(u0, sigma, nt)
```

```python
def plot(x, y):
    plt.figure()
    plt.plot(x,y)
    plt.xlabel('location')
    plt.ylabel('temperature')
    plt.grid()
plot(x, ans)
```

```python
# increase nt=1000
nt = 1000
ans2 = ftcs(u0, sigma, nt)
plot(x, ans2)
```

```python
# Increase the value of sigma to 1 to see that it is unstable
nt = 100
sigma = 1
```

```python
ans3 = ftcs(u0, sigma, nt)
plot(x, ans3)
```

```python
nt = 1000
sigma = 0.5
ans4 = ftcs_neumann(u0, sigma, nt)
plot(x, ans4)
```
